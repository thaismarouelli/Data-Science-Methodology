```python
import pandas as pd 

#import cvs file in the same folder of the jupter program 
data = pd.read_csv("credit.csv", index_col=0) 
```

```python
# type shows the type of any Python object 
type(data) 
```

```python
#print the first rows from the table 
data.head()
```

```python
#print the last rows from the table 
data.tail()
```

```python
#chaging the name of the index colum 
data.index.name = "Obs"
data.head()
```

```python
data.info() 
```

```python
#Selecting a column by label
print(data['Income'].head())
```

```python
#Selecting multiple columns by label
print(data[['Income','Education']].head())
```

```python
#another method to Selecting multiple columns by label
names=['Income','Education'] 
print(data[names].head())
```

```python
#Selecting a column by positional index
data.iloc[:,0].head()
```

```python
#Selecting multiple columns by positional indexing, the first argument delimits the rows and the second delimits the columns 
data.iloc[:,[0,5]].head()
```

```python
#Selecting rows by labels. The loc method alows one to select rows by the designated index labels (Obs).
data.loc[[1,2,5],:]
```

```python
# Unlike slicing with numerical indexing, slicing with label indexes includes the last item
data.loc[1:5,:]
```

```python
#Selecting rows by positional index
#This is useful when the index variable is a string or date.
data.iloc[10:20,3:5]
```

```python
#We can combine the previous examples to simultaneously select specific rows and columns.
data.loc[1:2,['Income', 'Education']] # label indexing
```

```python
#select columns by data types 
data.select_dtypes(["object"]).head()
```
