```python
import pandas as pd
data=pd.read_csv('Credit.csv',index_col=0)
data.index.name = "Obs"
```

```python
 print(data.info)
```

```python
# Dimension of dataset, i.e, rows and columns
print(data.shape)
```

```python
# Statistical summary of the dataset
print(data.describe()) 
```

```python
# To find the number of instances (rows) that belong to each class. Can be done for any nominal variable
print(data.groupby('Married').size())
```

```python
from matplotlib import pyplot as plt
# box and whisker plots
data.plot(kind='box', subplots=True, layout=(2,4), sharex=False, sharey=False)
plt.show()
```

```python
# histograms
data.hist()
plt.show()
```

```python
# histograms changing the parameters 
data.hist("Income")
plt.show()
```

```python
from pandas.plotting import scatter_matrix
# scatter plot matrix
scatter_matrix(data)
plt.show()
```

```python
married_counts = data["Married"].value_counts()
print(married_counts)
```

```python
married_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Distribution of Married Column")


plt.show()
```

```python
married_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Distribution of Married Column")

plt.xlabel("Married")
plt.ylabel("Count")
plt.show()
```

```python
ethnicity_counts = data["Ethnicity"].value_counts()
print(ethnicity_counts)   
```

```python
ethnicity_counts.plot(kind='bar', color=['skyblue', 'salmon', 'green'])
plt.title("Distribution of Ethnicity Column")

plt.xlabel("Ethnicity")
plt.ylabel("Count")
plt.show()
```
