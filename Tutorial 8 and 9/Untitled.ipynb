```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
import seaborn as sns  # another package for ploting


url = ("https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data")
abalone = pd.read_csv(url, header=None)
abalone.head() # To check the dataset
```

```python
# adding the columns names
abalone.columns = ["Sex","Length","Diameter","Height","Whole weight","Shucked weight", "Viscera weight",
"Shell weight", "Rings" ]
# Drop the sex column as we will not use it
abalone = abalone.drop("Sex", axis=1)
```

```python
# Statistical Analysis

import matplotlib.pyplot as plt
abalone["Rings"].hist(bins=15)
plt.show()
```

```python
#Checking the correlation of rings with other variables. A strong correlation between the independent and the goal variable 
#will be a good sign. 

correlation_matrix = abalone.corr()
correlation_matrix["Rings"]
```

```python
#Fitting a KKN Model; We will use this model to predict the age (by using the 
#number of rings) of the Abalone using the other physical features
X = abalone.drop("Rings", axis=1) # feature set
X = X.values
y = abalone["Rings"] # Label
y = y.values
```

```python
#Getting the training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12345)
knn_model = KNeighborsRegressor(n_neighbors=3) # KNN with K=3
knn_model.fit(X_train, y_train) # Fitting the model
```

```python
test_preds = knn_model.predict(X_test) #Making predictions
mse = mean_squared_error(y_test, test_preds) # Getting the error
rmse = sqrt(mse)
```

```python
#Plotting the Fit of the Model.
cmap = sns.cubehelix_palette(as_cmap=True)
f, ax = plt.subplots()
points = ax.scatter(
X_test[:, 0], X_test[:, 1], c=test_preds, s=50, cmap=cmap) # Using length and diameter
f.colorbar(points)
plt.xlabel('Length') 
plt.ylabel('Diameter') 
plt.show()
```

```python
import matplotlib.pyplot as plt 
from matplotlib import style
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs # For making dummy data

#Setting Plot stylesheet and generating dummy data
style.use("fivethirtyeight")
# make_blobs() is used to generate sample points around c centres (randomly chosen)
X, y = make_blobs(n_samples = 100, centers = 4, cluster_std = 1, n_features = 2)

#Plotting the data
plt.scatter(X[:, 0], X[:, 1], s = 30, color ='b')
# label the axes
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
plt.clf() # clear the figure
```

```python
import warnings
warnings.filterwarnings("ignore")

#Calculate the cost for different values of K
cost =[]
for i in range(1, 11):
    KM = KMeans(n_clusters = i, max_iter = 500)
    KM.fit(X)
# calculates the squared error for the clustered points
    cost.append(KM.inertia_) 

#Plot the cost against K values
plt.plot(range(1, 11), cost, color ='g', linewidth ='3')
plt.xlabel("Value of K")
plt.ylabel("Squared Error (Cost)")
plt.show() # clear the plot
```
